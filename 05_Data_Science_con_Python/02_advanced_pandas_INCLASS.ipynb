# Data Aggregation and Group Operations

## GroupBy mechanics

````python
df = pd.DataFrame({'key1' : ['a', 'a', 'b', 'b', 'a'],

          'key2' : ['one', 'two', 'one', 'two', 'one'],

          'data1' : np.random.randn(5),

          'data2' : np.random.randn(5)})

df
````

````python
grouped = df.groupby('key1')
````

````python
grouped.mean()
````

````python
grouped.count()['key2']
````

````python
grouped.agg(['mean', 'min', 'max'])
````

````python
grouped.agg({'data1': 'mean', 'data2': ['min', 'max']})
````

````python
grouped['data1'].agg(lambda series: series.max() - series.min())
````

````python
def span(series):
    return series.max() - series.min()
````

````python
grouped['data1'].agg(span)
````

### Iterating over groups

````python
for thing in df.groupby('key1'):
    print(type(thing))
````

````python
for key, group in df.groupby('key1'):
    print(key)
    print(group)
````

####  Exercise



What is the plane that has the highest average delay? We'll first group by `Tail_Number` (the *license plate* of a plane) and then calculate the relevant statistic for each group (group of *departures*)

* Get average delays for each plane, then show the top 10

* Get both average delay and number of flights. Show the 10 planes with the most flights

* Get both average delay and number of flights.  Show the 10 planes with the worst delays from those that had at least 10 flights.

````python
delays_per_plane = flights.groupby('Tail_Number')['DepDelay'].agg(['mean', 'count'])
````

````python
# average delays for each plane, then show the top 10

delays_per_plane.sort_values(by='mean', ascending=False).head(10)
````

````python
# average delay and number of flights. Show the 10 planes with the most flights

delays_per_plane.sort_values(by='count', ascending=False).head(10)
````

````python
# average delay and number of flights.  Show the 10 planes with the worst delays from those that had at least 10 flights.

delays_per_plane[delays_per_plane['count']>10].sort_values(by='mean', ascending=False).head(10)
````
