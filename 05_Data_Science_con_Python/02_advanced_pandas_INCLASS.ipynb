What is the plane that has the highest average delay? We'll first group by `Tail_Number` (the *license plate* of a plane) and then calculate the relevant statistic for each group (group of *departures*)

### Exercise: Filling missing values with group-specific values

we can fill missing values with specific group values

for instance, we want to fill missing distances with the mean of the flights of each plane, not the global mean

````python
def fillnas(df, column='data2'):
    return df.fillna({column: df[column].median()})
flights.groupby('Tail_Number').apply(fillnas, column='Distance')
````

# ONE HOT ENCODING

There is a function available in pandas to do a one hot encode but it is only usefull for categorical variables with few elements like for instance, the carrier of the flight.

For things like tail number this does not work due to the sheer number of features.

In this case we use target encoding.

HAY DEBERES

First we filter the dataframe so last day flights are not included.

Then, we calculate the mean grouping by Origin and Tail number

````python
flights_yesterday = flights[flights['FlightDate']<flights['FlightDate'].max()]

median_delay_origin = flights_yesterday.groupby('Origin')['DepDelay'].median()
median_delay_plane = flights_yesterday.groupby('Tail_Number')['DepDelay'].median()
````

After this we use map to encode a new variable (if we want to preserve the previous feature for whatever reason) with the median variables calculated.

Now we could start applying machine learning with these two new features.

````python
flights['median_delay_origin'] = flights['Origin'].map(median_delay_origin)
flights['median_delay_plane'] = flights['Tail_Number'].map(median_delay_plane)
````
