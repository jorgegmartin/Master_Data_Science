```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df = pd.read_csv('https://raw.githubusercontent.com/KaonToPion/datasets/main/room_occupancy.csv', parse_dates=True)
df['date']=pd.to_datetime(df['date'])
df=df.set_index('date')
df.head()
```

# Visualizar todas las series temporales en varios subplots y en un solo plot

```python
cols = df[1:]

for col in cols:
    fig, ax = plt.subplots(figsize=(8, 5))
    ax.plot(df.index, df[col], marker='.', linestyle='')
    ax.set_title(col,fontsize=8)
```

```python
fig, axs=plt.subplots(6, figsize=(12, 10), sharex=True)

for ax, col in zip(axs, cols):

  ax.plot(df.index, df[col])

  ax.set_title(col, fontsize=8)
```

```python
fig, ax = plt.subplots(figsize=(12, 8))
sns.lineplot(data=df, ax=ax)
```

```python
for col in cols:
  fig,ax = plt.subplots(figsize=(15,5))
  sns.scatterplot(data=df,x='date',y=col,hue='Occupancy',marker='.',ax=ax)
```
