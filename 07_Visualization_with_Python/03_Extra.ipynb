```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df = pd.read_csv('https://raw.githubusercontent.com/KaonToPion/datasets/main/room_occupancy.csv', parse_dates=True)
df.head()
```

```python
fig, ax = plt.subplots(6,figsize=(12, 8), sharex=True)
cols = df[1:]
for col in cols:
    fig, ax = plt.subplots(figsize=(12, 8))
    ax.plot(df['date'], df[col], marker='.', linestyle='')
```

```python
fig, ax = plt.subplots(figsize=(12, 8))
sns.lineplot(data=df, ax=ax)
```
